# Assignment 5: Semantic Relationships using WordNet

```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
def get_semantic_relationships(word):
    synonyms = set()
    antonyms = set()
    hypernyms = set()

    for synset in wn.synsets(word):
        # Synonyms
        for lemma in synset.lemmas():
            synonyms.add(lemma.name())

            # Antonyms
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

        # Hypernyms
        for hyper in synset.hypernyms():
            hypernyms.add(hyper.name().split('.')[0])

    return synonyms, antonyms, hypernyms
```

```python
word = "good"

synonyms, antonyms, hypernyms = get_semantic_relationships(word)

print("Word:", word)
print("\nSynonyms:")
print(synonyms)

print("\nAntonyms:")
print(antonyms)

print("\nHypernyms:")
print(hypernyms)
```

```python
text = "The movie was good and the acting was excellent"

words = text.lower().split()

for word in words:
    syn, ant, hyp = get_semantic_relationships(word)
    print(f"\nWord: {word}")
    print("Synonyms:", list(syn)[:5])
    print("Antonyms:", list(ant))
    print("Hypernyms:", list(hyp)[:5])
```
